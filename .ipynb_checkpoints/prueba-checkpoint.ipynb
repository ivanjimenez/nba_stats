## Parte 1: Construccin de la base de datos

### Discusin del modelo de datos

La tabla Excel que se nos presenta consta de las columnas (Team,Season, W, L, WIN%, etc. Las leyendas de estos atributos se ven ms abajo. En una primera observacin el modelo resultante de esta tabla podra realizarse de varias formas. En este documento voy a utilizar un modelo relacional para poder ser utilizado con SQL. Pero podemos utilizar otros noSQL con similar resultado.

Podemos formular la especificacin del modelo como en la NBA existen varios equipos los cuales participan cada uno en varias temporadas obteniendo unas estadsticas en cada una de estas temporadas. En cada temporada participan varios equipos. 

Como entidades podemos decir que estn los Equipos (Teams) y las Temporadas (Seasons), y que existe una relacin tiene_estadstica entre ambas.

Dicha relacin sera de N:M, siendo: Teams (1,n) <--->  Seasons (1,n). Para cada Equipo participa en una temporada y tiene varias estadsticas. Los atributos de cada entidad seran: Teams (idTeam(Primary Key), Name, Conference, ...) y Season (idSeason, Years,...). Se creara una relacin entre estas dos entidades. Season_team_stats(idTeam(FK,PK),idSeason(FK,PK),W,L,GP,WIN%,...). Teniendo esta una clave primaria compuesta de dos forneas para establecer la cardinalidad en el modelo relacional. El diagrama quedara:



Tambin se podra hacer una ternaria y que nos llevara a sacar una tabla por cada entidad: Teams, Seasons, Stats y luego la tabla de la relacin Equipo_tiene_estadsticas_en_temporada(id, idteams(fk), idseasons(fk), stats(fk)) y las tablal Teams(idteams, teamname, conference, ...), luego Seasons(idseason, years), y por ltimo Stats(idstat, W, L, W%, ...). A la hora de disear el modelo tendramos otra tabla mas adems que las bsquedas podran ser algo ms complejas.

Cabe decir que la Tabla Stats tambin podra ser dividida ya que podemos establecer estadsticas de tiro, rebote, etc. Pero eso sera una jerarquia que como opcin en el paso al modelo relacional podramos incluirlo todo en esta tabla. 

Voy a elegir realizar una relacin binaria y dejar stats como relacin con varios atributos. De esta forma saldran dos tabla para Teams y Seasons y una tercera tabla que sera Tiene_Estadsticas(idTeams(pk,fk), idSeason(pk,fk), W, L, W%...). Ahorras una tabla aunque tal vez pierdas algo de semntica, pero las queries pueden ser ms simples.

Se aporta el lenguaje DDL a continuacin adems del DML en la insercin. De esta forma tendramos la integridad referencial deseada.


Como ltima reflexin tambin podramos haber realizado **solo una tabla** como relacin sql y pasarla a DataFrame ya que con pandas se puede realizar incluso bsquecas muy potentes sin nisiqueira utilizar sql.


```sql
-- Estructura DDL

CREATE SCHEMA nba1;

CREATE TABLE nba1.`Teams` ( 
	`idTeams`            int  NOT NULL    PRIMARY KEY,
	`Name`               varchar(100)      ,
	`Conference`         enum(E,N,NE,W,S)      
 );

CREATE TABLE nba1.seasons ( 
	idseason             int  NOT NULL    PRIMARY KEY,
	year                 varchar(50)      
 );
 
-- El modelo de la tabla resultante de la relacin no tiene todos los atributos por comodidad de lectura

CREATE TABLE nba1.seasons_teams_stats ( 
	`Teams_ID`           int      ,
	`Seasons_ID`         int      ,
	gp                   decimal(3,2)      ,
	w                    decimal(3,2)      ,
	l                    decimal(3,2)      ,
	`win%`               decimal(3,5)      ,
	min                  decimal(3,2)      ,
	pts                  decimal(3,4)      ,
	fgm                  decimal(3,5)      ,
	fga                  decimal(3,5)      , 
	CONSTRAINT `Fk_Seasons_Teams1` FOREIGN KEY ( `Teams_ID` ) REFERENCES nba1.`Teams`( `idTeams` ) ON DELETE NO ACTION ON UPDATE NO ACTION,
	CONSTRAINT `Fk_Seasons_Teams2` FOREIGN KEY ( `Seasons_ID` ) REFERENCES nba1.seasons( idseason ) ON DELETE NO ACTION ON UPDATE NO ACTION
 );

CREATE TABLE nba1.stats ( 
	idstat               int  NOT NULL    PRIMARY KEY
 );
